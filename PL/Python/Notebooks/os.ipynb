```python
# because environ can changed between getting max length and print table
env = os.environ.copy()
max_length_environ_name = max(len(v_name) for v_name in env)
max_length_environ_value = max(len(value) for value in env.values())
type(max_length_environ_name)
```

```python
for variable, value in os.environ.items():
    print(
        f"{variable:{max_length_environ_name}}:\t{value:{max_length_environ_value}}\n"
        f"{'=' * 100}"
         )
```
